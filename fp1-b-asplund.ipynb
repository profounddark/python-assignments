```python
# fp1-b-lastname

rates = {'CAD': 1.2480163599, 'HKD': 7.7540286299, 'ISK': 125.3987730061, \
         'PHP': 48.6642126789, 'DKK': 6.0828629857, 'HUF': 294.6257668712, \
         'CZK': 21.3578732106, 'GBP': 0.7068139059, 'RON': 3.9875664622, \
         'SEK': 8.234601227, 'IDR': 14193.9468302658, 'INR': 72.6331288344, \
         'BRL': 5.453006135, 'RUB': 73.9074846626, 'HRK': 6.2081799591, \
         'JPY': 106.118609407, 'THB': 30.1496932515, 'CHF': 0.9060122699, \
         'EUR': 0.81799591, 'MYR': 4.0395092025, 'BGN': 1.5998364008, \
         'TRY': 7.2265030675, 'CNY': 6.4538241309, 'NOK': 8.3660531697, \
         'NZD': 1.3422494888, 'ZAR': 14.8139877301, 'USD': 1.0, \
         'MXN': 20.7299795501, 'SGD': 1.318609407, 'AUD': 1.2529243354, \
         'ILS': 3.2790184049, 'KRW': 1111.7709611452, 'PLN': 3.6909611452, \
         'USD': 1}


def isFloat(testValue):
    """Tests to see if a string can be converted to a float variable type

    Args:
        testValue (string): string to be tested

    Returns:
        bool: True if testValue can be converted to float, False otherwise
    """
    try:
        float(testValue)
        return True
    except:
        return False

def getCurrency():
    """Prompts the user to select a currency available in the rates dictionary.

    Returns:
        string: three letter code corresponding to the currency code selected by the user.
    """
    print("Available currencies are:")
    count = 0
    for currency, rate in rates.items():
        print(currency, end = " ")
        count += 1
        if (count % 6 == 0):
            print()
    print()
    validCode = False
    while (not validCode):
        print("Please enter target country code:")
        code = input()
        if (code in rates):
            validCode = True
        else:
            print("Country Code not found.")

    return code

def getAmount(code):
    """Prompts the user to enter an amount of USD to be converted to the target currency.

    Args:
        code (string): Three letter currency code to be converted to.

    Returns:
        float: Value of currency to be converted.
    """
    validAmount = False
    while (not validAmount):
        print("Enter the amount of USD to convert to " + code + ":")
        currencyAmount = input()
        if (isFloat(currencyAmount)):
            validAmount = True
        else:
            print("Invalid currency amount.")
    return float(currencyAmount)
    

 
currencyCode = getCurrency()
currencyAmount = getAmount(currencyCode)
print(f"Converting ${currencyAmount:.2f} to " + currencyCode + "...")

newCurrencyAmount = currencyAmount * rates[currencyCode]
print(f"You have converted {currencyAmount:.2f}USD to {newCurrencyAmount:.2f}" + currencyCode + ".")
```
