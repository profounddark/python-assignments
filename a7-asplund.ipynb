```python
# Constant for Seattle Tax Rate
SEA_TAX_RATE = 0.1035

class Restaurant:
    
    def __init__(self, name, menu):
        self.name = name
        self.menu = menu
        self.items_ordered = []
        self.managers = {"james":"bond", "lion":"king", "peter":"pan"}
        self.current_manager = ""

    def is_valid_login(self):
        """ is valid login:
            take input from user, id and pw
            check id exists in self.managers
            check pw matches key value in self.managers
            max 3 tries
            return True or False
            if Ture, set slef.current_manager if successful login
        """
        success = False

        username = input("User ID:")
        password = input("Password:")
        try:   
            if (self.managers[username] == password):
                print("Success!")
                self.current_manager = username
                success = True
        except:
            pass

        return success
        # write code here
            
    def print_menu(self):
        """ print menu after successful login:
        
            --SUISP Ordering System--
            Wlecome to, SHAKE SHOCK BURGER!
            Current manager, Lion.

            ------------
            The menu
            ------------
            [1] Shock Burger $4.85
            [2] French Fries $2.85
            [3] Milk Shake $4.5
        """
        print("__SUISP Ordering System--")
        print("Wlecome to, " + self.name)
        print("Current manager, " + self.current_manager)
        print("")
        print("------------\nThe menu\n------------")
        for item in range(len(self.menu)):
            print(f'[{item+1}] ' + self.menu[item][0] + ' ' + f' ${self.menu[item][1]:.2f}')


    def take_order(self, item, quantity):
        """ take order:
            item number and quantity
            update attribute self.items_ordered, i.e., {1:2, 3:1}
        """
        self.items_ordered.append([item, quantity])
    
    def process_order(self):
        """ process order:
            use attribute, items_ordered
            calculate sub total price on all items
            return sub total
        """
        sub_total = 0
        for order in self.items_ordered:
            sub_total += order[1] * self.menu[order[0]][1]

        return sub_total

    def print_sales_receipt(self):
        """ print sales receipt on this order:
            ++++++++++++++++++++++++
            The current order summary:
            ++++++++++++++++++++++++
            Shock Burger $4.85
                @ 3: item total $14.55
            French Fries $2.85
                @ 1: item total $2.85
            Milk Shake $4.50
                @ 5: item total $22.50
            ---------------------
            Sub Total: $39.90
            Tax: $3.99
            Total: $43.89
        """
        print("++++++++++++++++++++++++\nThe current order summary:\n++++++++++++++++++++++++")
        for order in self.items_ordered:
            print(self.menu[order[0]][0] + f' {self.menu[order[0]][1]:.2f}')
            cost = order[1] * self.menu[order[0]][1]
            print("    @ " + str(order[1]) + ": item total " + f'${cost:.2f}')
            
        print("---------------------")
        sub_total = self.process_order()
        tax = sub_total * SEA_TAX_RATE
        total = sub_total + tax
        
        print(f'Sub Total: ${sub_total:.2f}')
        print(f'Tax: ${tax:.2f}')
        print(f'Total: ${total:.2f}')


def main():
    menu = ["Shock Burger", float(4.85), 50], ["French Fries", float(2.85), 200], ["Milk Shake", float(4.50), 20]
    rt = Restaurant('Shake Shock Burger', menu)

    if rt.is_valid_login():
        rt.print_menu()
        more_items = "yes"
        while more_items == "yes":
            choice, quantity = int(input("\nWhat would you like?\n")), int(input("How many would you like?\n"))
            if 0 <= (choice-1) < len(rt.menu):
                rt.take_order(choice-1, quantity)
            more_items = (input("Do you want to order more items?")).lower()
        rt.print_sales_receipt()
    else:
        print("System closed due to incorrect tries!")

if __name__ == '__main__':
    main()
```
